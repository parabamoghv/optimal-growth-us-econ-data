```julia
using CSV, DataFrames, Plots
df_USA = CSV.read("../Data/PWT_USA_train.csv", DataFrame)
first(df_USA, 5)
```

# Curious about GDP expenditure and GDP output values.

At least for the US, they are similar

```julia
plot(
    df_USA.year, [df_USA.cgdpe df_USA.cgdpo],
    title="CGDPe vs CGDPo",
    label=["CGDPe" "CGDPo"],
    xlabel="Year", ylabel="Real GDP"
)

```

```julia
plot(
    df_USA.year, [df_USA.rgdpe df_USA.rgdpo],
    title="RGDPe vs RGDPo",
    label=["RGDPe" "RGDPo"],
    xlabel="Year", ylabel="Real GDP"
)
```

```julia
plot(
    df_USA.year, [df_USA.cgdpe_pc df_USA.cgdpo_pc],
    title="CGDPe Per Capita vs CGDPo Per Capita",
    label=["CGDPe" "CGDPo"],
    xlabel="Year", ylabel="Real GDP Per Capita"
)
```

# GDP per capita vs Capital Intensity (i.e. Capital Stock per capita)

This is indicative of Production function

```julia
scatter(
    df_USA.ck_pc, [df_USA.rgdpo_pc df_USA.rgdpe_pc],
    title="RGDP per capita vs Capital Intensity",
    label=["RGDPe" "RGDPo"],
    xlabel="Capital Intensity (C) per capita", ylabel="Real GDP Per Capita"
)
```

# Consumption vs GDP per capita

This is indicative of the policy function. We want analyize theorertical optimal policy function vs real observation

```julia
scatter(
    df_USA.rgdpo_pc, df_USA.ccon_pc,
    title="consumption per capita vs GDP per capita",
    label="consumption per capita",
    xlabel="Real GDP Per Capita", ylabel="Consumption Per Capita"
)
```

# Curious about the growth rates

```julia
# 100 .* diff(log.(df_USA.cgdpo))

plot(
    df_USA.year[2:end], 100 .* diff(log.(df_USA.cgdpo)),
    title="Log Growth Rate of CGDPo",
    label="Log Growth Rate of CGDPo",
    xlabel="Year", ylabel="log Growth Rate (%)"
)
```
